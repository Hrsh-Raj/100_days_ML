```python
df1_scale = df.copy()
df1_scale.head()
```

```python
def Z_Standardization(df1_scale, column):
    mean = df1_scale[column].mean()
    std = df1_scale[column].std()
    df1_scale[column] = (df1_scale[column] -mean)/std
    return df1_scale
```

```python
Z_Standardization(df1_scale, 'Age')
Z_Standardization(df1_scale, 'EstimatedSalary')
```
