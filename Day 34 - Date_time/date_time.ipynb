#### Working with Dates

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

##### 1) Extract year

```python
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

##### 2) Extract month

```python
date['date_month'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name']= date['date'].dt.month_name()
date.head()
```

##### 3) Extract date

```python
date['date_day_name'] = date['date'].dt.day_name()
```

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
#day of week
date['date_dow'] =date['date'].dt.dayofweek
date.head()
```

```python
#day of week - name
date['date_dow_name'] = date['date'].dt.day_name()
date.head()
```

```python
date2 = date.drop(columns=['product_id', 'city_id', 'orders', 'date_dow_name'])
date2.head()
```

```python
#is weekend?
date2['date_is_weekend'] = np.where(date2['date_day_name'].isin(['Sunday', 'Saturday']), 1, 0)
date2.head()
```

##### Extract week of the year

```python
date2['date_week'] = date2['date'].dt.weekday
date2.head()
```

```python
date2['date_week2'] = date2['date'].dt.isocalendar().week
date2.head()
```

##### Extract Quarter

```python
date2['quarter'] = date2['date'].dt.quarter
date2.head()
```

##### Extract Semester

```python
date2['semester'] = np.where(date2['quarter'].isin([1,2]),1,2)
date2.head()
```

##### Extract time elapsed between dates

```python
import datetime

today = datetime.datetime.today()
today
```

```python
today - date2['date']
```

```python
# to get the days
(today - date2['date']).dt.days
```
