### Bivariate and Mulivariate Analysis
#### 1) Scatterplot(Numerical - Numerical)

```python
#Bivariate
sns.scatterplot(x = 'total_bill', y ='tip', data = tips)
```

```python
#Plotting with color
#Multivariate
sns.scatterplot(x ='total_bill', y='tip', hue='smoker', data=tips)
plt.title('Total Bill')
plt.show()
```

```python
#For Linear Regression graph
sns.lmplot(x='total_bill', y ='tip', hue='smoker', data = tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data = tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=tips)
```

#### 2) Bar Plot(Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Age', data=titanic)
#Black tells us about confidence interval
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic)
```

```python
sns.barplot(x='Pclass', y='Age', hue='Sex', data=titanic)
```

#### 3) Box plot(Numerical-Categorical)

```python
sns.boxplot(x='Sex', y='Age', data=titanic)
```

```python
sns.boxplot(x='Age', y='Sex', hue='Survived', data=titanic)
```

#### 4) Distplot(Numerical-Categorical)

```python
#We have used Probability Density function in distplot at the time of univarate, we can do the same here too.
sns.distplot(titanic['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
sns.distplot(titanic[titanic['Survived']==1]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

#### 5) HeatMap(Categorical - Categorical)

```python
titanic.head(3)
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean()*100
```

```python
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100)
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```

```python
(titanic.groupby('Embarked')['Survived'].mean()*100).plot(kind='bar')
```
